```python
def print_unique_elements(_df: pd.DataFrame) -> dict:
    """
    [input]
        _df : [pandas.DataFrame] read_statsdataで読み込んだテーブル
    [output]
        None
    """
    cols = [col for col in _df.columns if col != "value"]
    for _col in cols:
        print(f"{_col} : ", _df[_col].unique())
```
